## 3.1 KNN Clustering Example 1 (YEAST)

```python
import scipy.io
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import warnings
```

```python
path = './yeastData310.mat'
data = scipy.io.loadmat(path)
X = data['X']
```

```python
# Cluster yeast data using Kmeans

kmeans = KMeans(n_clusters=16, n_init=10, random_state=0, algorithm='lloyd').fit(X)
times = data['times']
X = X.transpose()
```

```python
labels = kmeans.labels_ #whats in here?
clu_cen = kmeans.cluster_centers_
```

```python
clusters = [[] for i in range(0, 16)]

for (i, l) in enumerate(labels):
    clusters[l].append(i)

times = times.reshape((7,))
```

```python
#Visualizing all the time series assigned to each cluster

plt.figure()
for l in range(0, 16):
    plt.subplot(4, 4, l + 1)
    if clusters[l] != []:
        plt.plot(times, X[:, clusters[l]])
plt.suptitle("K-Means Clustering of Profiles")
plt.show()
```

```python
# Visualizing the 16 cluster centers as prototypical time series.

plt.figure()
for l in range(0, 16):
    plt.subplot(4, 4, l + 1).axis('off')
    plt.plot(times, clu_cen[l, :])
plt.suptitle("K-Means centroids")
plt.show()
```
